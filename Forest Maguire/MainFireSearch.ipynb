```python
import pandas as pd
import numpy as np
import geopandas
#import folium
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

```

```python
dat_in = pd.read_csv("FireDataSet.csv")
mF = pd.DataFrame(dat_in)

#mF.head(1)
#betweet = mF.loc[:, ("incident_type" , "alarms","neighborhood")]
#mf_sub = betweet.dropna()
#print(mf_sub.head(10))
```

```python
#mF["incident_type"].hist(bins = 10)
comp_dat = mF.groupby("neighborhood").sum()["alarms"].sort_values(ascending = True)

sum_dat = mF.groupby("type_description").sum()["alarms"].sort_values(ascending = True)
```

```python

Alarm_Fire_graph = comp_dat.plot.barh(figsize=(21 , 30), color = 'blue')
plt.ylabel('Neighborhood',fontweight='heavy', color = 'gold', fontsize='25', horizontalalignment='center')
plt.xlabel('Alarms Responses' , fontweight='bold', color = 'gold' , fontsize='18', horizontalalignment = 'center')
plt.yticks(fontsize=20, color = 'navy')
plt.xticks(fontsize=15, color = 'navy')

Alarm_Fire_graph.set_facecolor('pink')
Alarm_Fire_graph.invert_yaxis()

```

```python
#print(sum_dat.tail(10))
type_graph = sum_dat.plot.barh(figsize=(20,20))
plt.xticks(fontsize = .5)
plt.ylabel("Alarm types")
```

```python
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
```

```python
fire_map = neighborhoods.merge(comp_dat, how='left', left_on='hood', right_on='neighborhood')

fire_map[['objectid','alarms','geometry']].head()


matplotlib.rcParams['font.size'] = 40

newmap = fire_map.plot(column = "alarms",
              cmap = 'spring_r', 
              edgecolor = 'black', 
              figsize =(50,50),
              legend = True, 
              legend_kwds ={'label': "Amount of Fire Depardment Responses", 'orientation' : 'horizontal'},)


newmap.set_facecolor('lightblue')

plt.yticks([])
plt.xticks([])
```
